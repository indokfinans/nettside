```python
import yfinance as yf

ticker_symbols = ['ABL.OL', 'AKRBP.OL', 'BRG.OL', 'CRAYN.OL', 'DNB.OL', 'GJF.OL', 'MOWI.OL', 'NKR.OL', 'TEL.OL', 'VAR.OL', 'MGN.OL', 'DVD.OL']
# get last month's data
data = yf.download(ticker_symbols, period='1mo')
```

```python
data
```

```python
ticker_symbols = [symbol.strip().upper() for symbol in ticker_symbols]
ticker_symbols
```

```python

for symbol in ticker_symbols:
    last_price = data['Close'][symbol].iloc[-1]  # Last available closing price
    print(f'{symbol}: {last_price:.2f} NOK')
```
